```
with open("test1.txt","w") as file:
  file.write("abc")

converter = tf.lite.TFLiteConverter.from_keras_model_file( "NASNetMobile.h5" ) 
model = converter.convert()
file = open( 'NASNetMobile.tflite' , 'wb' ) 
file.write(model)
file.close()
```
